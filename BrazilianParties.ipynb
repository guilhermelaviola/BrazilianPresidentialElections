```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from google.colab import drive
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
drive.mount('/content/drive')
```

```python
# Setting the search parameters:
filename = 'presidential-elections-br-totals.xlsx'
search_path = '/content/drive/My Drive/Datasets/Elections'
# Search for the file
file = None
for root, dirs, files in os.walk(search_path):
    if filename in files:
        file= os.path.join(root, filename)
        break
    df = pd.read_excel(file)
```

```python
# Importing the 'Total votes' dataset:
df = pd.ExcelFile(file)
parties = pd.read_excel(df, 'Parties')
```

```python
# Displaying the last 10 parties:
parties.tail(10)
```

```python
# Displaying the most present ideologies:
parties['Ideology'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Most present ideologies', fontweight = 'bold')
plt.show()
```

```python
# Displaying the most present orientations:
parties['Orientation'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Most present orientations', fontweight = 'bold')
plt.show()
```
